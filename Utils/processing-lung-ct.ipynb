# Split dataset betwen train and validation

```python
import json

# Specify the path to the JSON file
json_path = '/content/LungCT/LungCT_dataset.json'

# Open and load the JSON file
with open(json_path, 'r') as json_file:
    json_data = json.load(json_file)
```

```python
train = [image['image'].replace('./imagesTr/', '/content/LungCT/imagesTr/') 
         for image in json_data['training']]
train = sorted(train)
```

```python
validate = [image['image'].replace('./imagesTs/', '/content/LungCT/imagesTs/') 
         for image in json_data['test']]
validate = sorted(validate)
```

# Plot Train

```python
for i in range(0, len(train)):
  nii_img = nib.load(train[i]).get_fdata()
  nii_label = nib.load(train[i].replace('/imagesTr/', '/masksTr/')).get_fdata()

  print(nii_img.shape, train[i])
  print(nii_label.shape, train[i].replace('/imagesTr/', '/masksTr/'))

  # Create a figure with two subplots
  fig, axes = plt.subplots(1, 2)

  # Plot the middle slice of image in the first subplot
  axes[0].imshow(nii_img[:, :, nii_img.shape[2] // 2] / 255.0, cmap='gray')
  axes[0].axis('off')
  axes[0].set_title('Image')

  # Plot the second slice in the second subplot
  axes[1].imshow(nii_label[:, :, nii_label.shape[2] // 2] / 255.0, cmap='gray')
  axes[1].axis('off')
  axes[1].set_title('Label')

  # Adjust the spacing between subplots
  plt.subplots_adjust(wspace=0.3)

  # Display the figure
  plt.show()

  print()
```

# Plot Validate

```python
for i in range(0, len(validate)):
  nii_img = nib.load(validate[i].replace('.csv', '.nii.gz')).get_fdata()
  nii_label = nib.load(validate[i]
                       .replace('/imagesTs/', '/masksTs/')
                       .replace('.csv', '.nii.gz')).get_fdata()

  print(nii_img.shape, validate[i])
  print(nii_label.shape, validate[i].replace('/imagesTr/', '/masksTr/'))

  # Create a figure with two subplots
  fig, axes = plt.subplots(1, 2)

  # Plot the middle slice of image in the first subplot
  axes[0].imshow(nii_img[:, :, nii_img.shape[2] // 2] / 255.0, cmap='gray')
  axes[0].axis('off')
  axes[0].set_title('Image')

  # Plot the second slice in the second subplot
  axes[1].imshow(nii_label[:, :, nii_label.shape[2] // 2] / 255.0, cmap='gray')
  axes[1].axis('off')
  axes[1].set_title('Label')

  # Adjust the spacing between subplots
  plt.subplots_adjust(wspace=0.3)

  # Display the figure
  plt.show()

  print()
```

# Processing the dataset

```python
for image in train:
  file_name = image.replace('/content/LungCT/imagesTr/', '')
  directory = f'Processed_LungCT/train/{file_name.replace(".nii.gz", "")}'

  if not os.path.exists(directory):
    os.makedirs(directory)

  nii_img = nib.load(image).get_fdata()
  nii_mask = nib.load(image.replace('/imagesTr/', '/masksTr/')).get_fdata()

  # Create a NIfTI image object from the NumPy array
  nifti_img = nib.Nifti1Image(nii_img[:, :, nii_img.shape[2] // 2] / 255.0, affine=np.eye(4))
  nifti_mask = nib.Nifti1Image(nii_mask[:, :, nii_mask.shape[2] // 2] / 255.0, affine=np.eye(4))

  # Save the NIfTI image to a compressed NIfTI file
  nib.save(nifti_img, f'Processed_LungCT/train/{file_name.replace(".nii.gz", "")}/slice_middle.nii.gz')
  nib.save(nifti_mask, f'Processed_LungCT/train/{file_name.replace(".nii.gz", "")}/slice_middle_mask.nii.gz')
```

```python
for image in validate:
  file_name = image.replace('/content/LungCT/imagesTs/', '')
  directory = f'Processed_LungCT/validate/{file_name.replace(".csv", "")}'
  if not os.path.exists(directory):
    os.makedirs(directory)
  nii_img = nib.load(image
                     .replace('.csv', '.nii.gz')).get_fdata()
  nii_mask = nib.load(image
                      .replace('.csv', '.nii.gz')
                      .replace('/imagesTs/', '/masksTs/')).get_fdata()

  # Create a NIfTI image object from the NumPy array
  nifti_img = nib.Nifti1Image(nii_img[:, :, nii_img.shape[2] // 2] / 255.0, affine=np.eye(4))
  nifti_mask = nib.Nifti1Image(nii_mask[:, :, nii_mask.shape[2] // 2] / 255.0, affine=np.eye(4))

  # Save the NIfTI image to a compressed NIfTI file
  nib.save(nifti_img, f'Processed_LungCT/validate/{file_name.replace(".csv", "")}/slice_middle.nii.gz')
  nib.save(nifti_mask, f'Processed_LungCT/validate/{file_name.replace(".csv", "")}/slice_middle_mask.nii.gz')
```
