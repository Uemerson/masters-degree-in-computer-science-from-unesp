# Split dataset between train and validate

```python
import pathlib
path = pathlib.Path('/content/OASIS')
list_images = [str(f/'slice_norm.nii.gz') for f in sorted(path.iterdir()) if str(f).endswith('MR1')]

with open('list.txt','w') as tfile:
	tfile.write('\n'.join(list_images))
```

```python
random_images = sorted(random.sample(list_images, int(0.8 * len(list_images))))
train = random_images
validate = sorted([item for item in list_images if item not in random_images])
```

```python
for element in train:
  if element in validate:
    print("There is one element in common:: ", element)
    break
else:
  print("There are no elements in common between the lists.")
```

```python
with open('train.txt','w') as tfile:
	tfile.write('\n'.join(train))
```

```python
with open('validate.txt','w') as tfile:
	tfile.write('\n'.join(validate))
```
